````python
# check other columns
kolom_temp = ['AMT_REQ_CREDIT_BUREAU_DAY',
              'AMT_REQ_CREDIT_BUREAU_HOUR',
              'AMT_REQ_CREDIT_BUREAU_MON',
              'AMT_REQ_CREDIT_BUREAU_QRT',
              'AMT_REQ_CREDIT_BUREAU_WEEK',
              'AMT_REQ_CREDIT_BUREAU_YEAR',
              'CNT_FAM_MEMBERS', 
              'AMT_ANNUITY',
              'AMT_GOODS_PRICE',
              'DEF_30_CNT_SOCIAL_CIRCLE',
              'DEF_60_CNT_SOCIAL_CIRCLE',
              'NAME_TYPE_SUITE',
              'OBS_30_CNT_SOCIAL_CIRCLE',
              'OBS_60_CNT_SOCIAL_CIRCLE',
              'OCCUPATION_TYPE']

for i in kolom_temp:
    if data_train[i].nunique() < 35:
        # untuk kategorikal data
        mode = data_train[i].mode()[0]
        data_train[i].fillna(mode, inplace=True)
    else:
        # untuk numerikal data
        mean = data_train[i].mean()
        data_train[i].fillna(mean, inplace=True)
````

````python
def bar_chart(df, column_name):
    data = df.groupby(column_name)['TARGET'].value_counts().to_frame('counts').reset_index()
    data = data.pivot(index=column_name, columns='TARGET', values='counts').sort_values(by=column_name)
    data = data.rename(columns={0: 'Paid', 1:'Unpaid'})

    persen = df[['TARGET', column_name]].groupby(column_name).aggregate(['count', 'mean'])
    
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(18,5))
    data.plot(kind='bar', stacked=True, color=colors, ax=ax[0], sort_columns=True)
    
    dt = data_clean[['TARGET', column_name]].groupby(column_name).aggregate(['count', 'mean'])
    dt[('TARGET',  'mean')].plot(kind='bar', stacked=False, color='blue', ax=ax[1], sort_columns=True)
    
    fig.suptitle(column_name, fontsize=16)
````

````python
categorical0_5= categorical[0:5]

for column_name in categorical0_5:
    bar_chart(data_clean, column_name)
````

Meskipun jumlah gagal bayar pada nilai tersebut paling tinggi, namun persentase gagal bayar tidak terlalu tinggi (dibandingkan dengan nilai lainnya pada kategori tersebut). Hal ini disebabkan oleh jumlah data yang sangat tidak merata, maka harus kita lihat lebih detail, dan tidak terpacu hanya dari persentase gagal bayarnya saja.
